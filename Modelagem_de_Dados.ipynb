# Checando Estações

```python
def estacao(mes):
  if mes >= 9 and mes <= 11:
    return 'Primavera'
  elif mes == 12 or mes <= 2:
    return 'Verão'
  elif mes >= 3 and mes <= 5:
    return 'Outono'
  else:
    return 'Inverno'
```

```python
df['Estação'] = df.Month.apply(estacao)
df.head()
```

```python
df_medias_estacao = df.groupby(by=['Year', 'Estação']).mean()
df_medias_estacao
```

# Análise de Regiões Climáticas

[Regiões Climáticas da Austrália](https://australianwildlifejourneys.com/plan-your-trip/climate-and-seasons)

[Imagem da Austrália](https://australia101.com/uploads/images/australia-climate-map.png)

```python
def regiao(location):
  if location == 'Canberra' or location == 'Tuggeranong' or location == 'MountGinini':
    return 'Temperado'    # ACT
  if location == 'Sydney' or location == 'Penrith' or location == 'Newcastle' or location == 'Wollongong' or location == 'SydneyAirport' or location == 'BadgerysCreek' or location == 'Cobar' or location == 'CoffsHarbour' or location == 'Moree' or location == 'NorahHead' or location == 'NorfolkIsland' or location == 'Richmond' or location == 'WaggaWagga' or location == 'Williamtown':
    return 'Temperado'  # NSW
  if location == 'Melbourne' or location == 'Ballarat' or location == 'Albury' or location == 'Bendigo' or location == 'MelbourneAirport' or location == 'Sale' or location == 'Mildura' or location == 'Nhil' or location == 'Portland' or location == 'Watsonia' or location == 'Dartmoor':
    return 'Temperado'   # Victoria
  if location == 'Brisbane' or location == 'Cairns' or location == 'Townsville' or location == 'GoldCoast':
    return 'Subtropical'   # Queensland
  if location == 'Adelaide' or location == 'MountGambier' or location == 'Nuriootpa' or location == 'Woomera':
    return 'Deserto'   # SA
  if location == 'Perth' or location == 'Albany' or location == 'Witchcliffe' or location == 'PerthAirport' or location == 'PearceRAAF' or location == 'SalmonGums' or location == 'Walpole':
    return 'Pradaria'   # WA - Grassland
  if location == 'Hobart' or location == 'Launceston':
    return 'Temperado'   # Tasmania
  if location == 'Darwin' or location == 'AliceSprings' or location == 'Katherine' or location == 'Uluru':
    return 'Tropical'   # NT
```

```python
df['Região'] = df.Location.apply(regiao)
df.head()
```

```python
df['Região'].unique()
```

```python
df_medias_estacao = df.groupby(by=['Year', 'Estação', 'Região']).mean()
df_medias_estacao.head(20)
```

# Substituindo Valores Nulos pelas Médias

```python
df_agrupado = df.set_index(['Year', 'Estação', 'Região'])
df_agrupado.head()
```

```python
df_merge = df_agrupado.merge(df_medias_estacao[['Evaporation', 'Sunshine', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 
                                                'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']], how = 'left', left_index = True, right_index = True)
df_merge.head()
```

```python
df_merge = df_merge.reset_index()
df_merge.head()
```

```python
import numpy as np
```

```python
# Caso a condição seja True, pega o primeiro valor (Evaporation_y é da tabela de médias)
df_merge['Evaporation_x'] = np.where(df_merge['Evaporation_x'].isnull(), df_merge['Evaporation_y'], df_merge['Evaporation_x'])
df_merge['Humidity3pm_x'] = np.where(df_merge['Humidity3pm_x'].isnull(), df_merge['Humidity3pm_y'], df_merge['Humidity3pm_x'])
df_merge['Humidity9am_x'] = np.where(df_merge['Humidity9am_x'].isnull(), df_merge['Humidity9am_y'], df_merge['Humidity9am_x'])
df_merge['Pressure3pm_x'] = np.where(df_merge['Pressure3pm_x'].isnull(), df_merge['Pressure3pm_y'], df_merge['Pressure3pm_x'])
df_merge['Pressure9am_x'] = np.where(df_merge['Pressure9am_x'].isnull(), df_merge['Pressure9am_y'], df_merge['Pressure9am_x'])
df_merge['Sunshine_x'] = np.where(df_merge['Sunshine_x'].isnull(), df_merge['Sunshine_y'], df_merge['Sunshine_x'])
df_merge['Temp3pm_x'] = np.where(df_merge['Temp3pm_x'].isnull(), df_merge['Temp3pm_y'], df_merge['Temp3pm_x'])
df_merge['Temp9am_x'] = np.where(df_merge['Temp9am_x'].isnull(), df_merge['Temp9am_y'], df_merge['Temp9am_x'])
df_merge['WindSpeed3pm_x'] = np.where(df_merge['WindSpeed3pm_x'].isnull(), df_merge['WindSpeed3pm_y'], df_merge['WindSpeed3pm_x'])
df_merge['WindSpeed9am_x'] = np.where(df_merge['WindSpeed9am_x'].isnull(), df_merge['WindSpeed9am_y'], df_merge['WindSpeed9am_x'])
df_merge['Cloud3pm_x'] = np.where(df_merge['Cloud3pm_x'].isnull(), df_merge['Cloud3pm_y'], df_merge['Cloud3pm_x'])
df_merge['Cloud9am_x'] = np.where(df_merge['Cloud9am_x'].isnull(), df_merge['Cloud9am_y'], df_merge['Cloud9am_x'])
```
