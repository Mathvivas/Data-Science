## Valor Médio Total da Compra pelo Número de Parcelas

```python
df_payments.head(2)
```

```python
df_payments.groupby(by='payment_installments').mean()
```

```python
df_plot = df_payments.groupby(by='payment_installments').mean()
fig = px.bar(df_plot, x=df_plot.index, y='payment_value',
             title="Valor Médio Total da Compra pelo Número de Parcelas",
             labels={
                 'index':'Número de Parcelas',
                 'payment_value':'Valor Médio da Compra'
             })
fig.show()
```

```python
df_pay_parc = df_payments.groupby(by='payment_installments').mean().reset_index().query('payment_installments > 0')
df_pay_parc['v_por_parc'] = df_pay_parc.payment_value / df_pay_parc.payment_installments
df_pay_parc
```

```python
import plotly.graph_objects as go
x = df_pay_parc.payment_installments

fig = go.Figure(data = [
            go.Bar(name = 'Valor Total da Compra', x = x, y = df_pay_parc.payment_value),
            go.Bar(name = 'Valor da Parcela', x = x, y = df_pay_parc.v_por_parc)
])
# Change the Bar Mode
fig.update_layout(barmode = 'group')
fig.show()
```

## Análise sobre Atrasos

```python
col_temp = df_orders.order_estimated_delivery_date - df_orders.order_delivered_customer_date
df_orders['order_delay'] = col_temp.dt.days
df_orders.head()
```

```python
df_plot = df_orders.order_delay <= 0
df_plot.value_counts()
```

```python
df_plot = df_plot.value_counts()
fig = px.pie(df_plot, values = 'order_delay', names = ['Em dia', 'Atrasado'])
fig.show()
```

```python
df_aux_cust_order = df_orders.merge(df_customers, how='inner', on='customer_id')
df_aux_cust_order.head()
```

```python
df_aux_cust_order_item = df_aux_cust_order.merge(df_order_items, how='inner', on='order_id')
df_aux_cust_order_item.head()
```

```python
df_aux_cust_order_item_product = df_aux_cust_order_item.merge(df_products, how='inner', on='product_id')
df_aux_cust_order_item_product.head()
```

```python
df_mais_vendidos = df_aux_cust_order_item_product[['customer_state', 
                      'product_category_name', 'price']].groupby(by=['customer_state', 
                                                                    'product_category_name']).agg(['count', 'sum'])
df_mais_vendidos
```

```python
df_SP_RJ_AM = df_mais_vendidos.query('customer_state in ["SP", "RJ", "AM"]')
df_SP_RJ_AM
```

```python
df_SP_RJ_AM.sort_values(by=['customer_state', ('price', 'sum')], ascending=False).reset_index()
```

```python
SP = df_SP_RJ_AM.query('customer_state == "SP"').sort_values(by=['customer_state', ('price','sum')], ascending= False).head(5)
SP
```

```python
RJ = df_SP_RJ_AM.query('customer_state == "RJ"').sort_values(by=['customer_state', ('price','sum')], ascending= False).head(5)
RJ
```

```python
AM = df_SP_RJ_AM.query('customer_state == "AM"').sort_values(by=['customer_state', ('price','sum')], ascending= False).head(5)
AM
```
