## Analisando a Média dos Dias da Semana

```python
temp_por_dia_semana = df_tempo.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temp_por_dia_semana = temp_por_dia_semana[nome_dias]
temp_por_dia_semana
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temp_por_dia_semana))

eixo.bar(indice, temp_por_dia_semana, color='yellow', edgecolor='k')
eixo.set_title('Temperatura por Dia da Semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da Semana', fontsize=15)
eixo.set_ylabel('Temperatura Média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

## Analisando Pela Pizza

```python
fig = plt.figure(figsize=(12, 10))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.pie(temp_por_dia_semana, labels=temp_por_dia_semana.index, autopct='%.1f%%')
eixo.set_title('Temperatura por Dia da Semana', fontsize=15, pad=10)
```

### Explodindo a Pizza

```python
fig = plt.figure(figsize=(12, 10))
eixo = fig.add_axes([0, 0, 1, 1])

explodir = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temp_por_dia_semana, labels=temp_por_dia_semana.index, autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por Dia da Semana', fontsize=15, pad=10)
```

## Colorindo as Espécies Diferentes

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}

for especie in df_iris['espécie'].unique():
  tmp = df_iris[df_iris['espécie'] == especie]
  eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color=cores[especie])

eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da Sétala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

## Unindo os Gráficos em uma Mesma Imagem

```python
from PIL import Image
```

```python
largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new('RGB', (3 * largura, altura))
intervalo = 0

for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png', 'histograma_ivg.png']):
  combinada.paste(imagem, (intervalo, 0))
  intervalo += largura

combinada.save('combinada.png')
```
